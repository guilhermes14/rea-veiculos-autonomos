```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Valor real / Cenário perfeito
VR= 10.0
```

 *U* representa média de erro e *S* representa desvio padrão

```python
def comparador_de_sensores(VR, u1, s1, u2, s2):
  #Sensor 1
  Erro_1= np.random.normal(u1,s1,5000)
  Sensor_1= VR+Erro_1

  #Sensor 2
  Erro_2= np.random.normal(u2,s2,5000)
  Sensor_2= VR+Erro_2

  # Análise de gráficos dos sensores
  plt.plot(Sensor_1)
  plt.title('Medidas Sensor 1')
  plt.xlabel('n')
  plt.ylabel('Valor')
  plt.grid()
  plt.show()
  plt.plot(Sensor_2)
  plt.title('Medidas Sensor 2')
  plt.xlabel('n')
  plt.ylabel('Valor')
  plt.grid()
  plt.show()

  #Histograma dos sensores
  plt.hist(Sensor_1,100) # 10,100
  plt.title('Histograma Sensor 1')
  plt.xlabel('Medidas')
  plt.ylabel('Número de ocorrências')
  plt.grid()
  plt.show()
  plt.hist(Sensor_2,100) # 10,100
  plt.title('Histograma Sensor 2')
  plt.xlabel('Medidas')
  plt.ylabel('Número de ocorrências')
  plt.grid()
  plt.show()

  # Fixando limites eixo X
  plt.hist(Sensor_1,100,[-10,30])
  plt.title('Histograma Sensor 1')
  plt.xlabel('Medidas')
  plt.ylabel('Número de ocorrências')
  plt.grid()
  plt.show()
  plt.hist(Sensor_2,100,[-10,30]) # 10,100
  plt.title('Histograma Sensor 2')
  plt.xlabel('Medidas')
  plt.ylabel('Número de ocorrências')
  plt.grid()
  plt.show()

  med_1= np.mean(Sensor_1)
  dev_pad_1= np.std(Sensor_1)

  med_2= np.mean(Sensor_2)
  dev_pad_2= np.std(Sensor_2)

  print('med_1=',med_1)
  print('med_2=',med_2)
  print('-----------------------------------')
  print('dev_pad_1=',dev_pad_1)
  print('dev_pad_2=',dev_pad_2)
  print('-----------------------------------')

  ES_1= np.abs(VR-med_1)
  ES_2= np.abs(VR-med_2)
  print('ES_1=',ES_1)
  print('ES_2=',ES_2)

  erro_total1 = ES_1 + dev_pad_1
  erro_total2 = ES_2 + dev_pad_2
  print('-----------------------------------')
  print('Erro Total Sensor 1=', erro_total1)
  print('Erro Total Sensor 2=', erro_total2)
  print('-----------------------------------')

  if erro_total1 < erro_total2:
    print('Sensor 1 é melhor')
  elif erro_total2 < erro_total1:
    print('Sensor 2 é melhor')
  else:
    print('Os sensores têm desempenho similar')
```

#Configuração 1

```python
comparador_de_sensores(VR, 1, 1, 1, 5)
```

```python
comparador_de_sensores(VR, 1, 5, 1, 1)
```
